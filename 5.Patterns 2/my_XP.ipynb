1.Inverted Triangle-

```python
n = int(input())
i = 1

while i<=n:
    j = 1
    while j<=n-i+1:
        print("*",end="")
        j= j+1
    print()
    i = i+1
```

2.Code %3a Inverted Number Pattern-

```python
n = int(input())
i = 1

while i<=n:
    j = 1
    while j<=n-i+1:
        print(i ,end="")
        j= j+1
    print()
    i = i+1
```

3.Reversed Pattern-

```python
n = int(input())
i = 1

while i<=n:
    spaces=1
    while spaces <= n-i:
        print(" ",end=" ")
        spaces=spaces+1
    stars = 1
    while stars <= i:
        print("*",end = " ")
        stars = stars+1
    print()
    i = i+1
```

4.Code %3a Mirror Number Pattern-

```python
n = int(input())
i = 1

while i<=n:
    spaces=1
    while spaces <= n-i:
        print(" ",end=" ")
        spaces=spaces+1
    j = 1
    while j <= i:
        print(j,end = " ")
        j = j+1
    
    print()
    i = i+1
```

5.Isosceles Pattern-

```python
n = int(input())
i =1 

#for space
while i<= n:
    spaces = 1
    while spaces <= n-i:
        print(" ",end = " ")
        spaces = spaces+1
 #for print       
    p = 1
    j = 1
    #For Incresing seq
    while j <= i:
        print(p ,end=" ")
        j = j+1
        p = p+1
    #For decresing seq
    p = i - 1
    while p>=1:
        print(p,end=" ")
        p = p -1
    print()
    i = i+1
```
