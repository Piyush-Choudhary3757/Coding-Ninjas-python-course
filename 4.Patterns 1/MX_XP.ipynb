4.Square Patterns-

```python
n = int(input())
i = 1
while i<=n:
    j=1
    while j<=n:
        print(j,end=" ")
        j=j+1
    print()
    i=i+1
```

```python
n = int(input())
i = 1
while i<=n:
    j=1
    while j<=n:
        print(n-j+1,end=" ")
        j=j+1
    print()
    i=i+1
```

5.Triangular Patterns-

```python
n = int(input())
i = 1
while i<=n:
    j=1
    while j<=i:
        print(j,end=" ")
        j=j+1
    print()
    i=i+1
```

```python
#with formula
n = int(input())
i = 1

while i<=n:
    j=1
    while j<=i:
        print(j+i-1,end=" ")
        j=j+1
    print()
    i=i+1
```

```python
#using a p

n = int(input())
i = 1
while i<=n:
    j = 1
    p = 1
    while j<=i:
        print(p,end=" ")
        j=j+1
        p=p+1
    print()
    i=i+1
```

```python
n = int(input())
i = 1

while i<=n:
    j=1
    while j<=i:
        #print jth Column
        charP = chr(ord("A") + j-1)
        print(charP,end =" ")
        j=j+1
    print()
    i=i+1
```

```python
n = int(input())
i = 1

while i<=n:
    j=1
    start_char = chr(ord("A") + j-1)
    while j<=i:
        #print jth Column
        charP = chr(ord(start_char) + j-1)
        print(charP ,end =" ")
        j=j+1
    print()
    i=i+1
```
